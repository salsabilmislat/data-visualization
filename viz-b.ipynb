```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## gitting the datasets from seaborn

```python
sns.set()
sns.get_dataset_names()
```

## Load the data receive into a Pandas DataFrame

```python
titanic= sns.load_dataset('titanic')
titanic= pd.DataFrame(titanic)
titanic
```

## Showing the first 20 rows of the data set.

```python
titanic.head(20)
```

## Analyzing the penguins dataset

### Visualize the relationship for the people who get paid (fare) for the male and female depending on the classifiction for the pepole. we can notice that the fisrt class get paid the most hight price while the other classes git paid less than the fisrt class

```python
sns.set_theme(style="white")
sns.relplot(x='class', y='fare', hue="sex", data=titanic)
```

### Visualize the relationship for the pepole who survived or not surviveddepending on the gender.we can notice that the female has the large Percentage of survived, while the male Percentage is much lesser than the female 

```python
sns.displot(data=titanic, x="sex", hue="survived", col="survived")
```

### Visualize the relationship for the female who get paid (fare) depending on the classifiction for the pepole.we can notice that the fisrt class get paid the most hight price while the other classes git paid less than the fisrt class

```python
female_data = titanic [ titanic['sex'] == 'female' ]
female_data
sns.set_theme(style="white")
sns.relplot(x='class', y='fare', hue="sex", data=female_data)
```
